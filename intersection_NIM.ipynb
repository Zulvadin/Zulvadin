```
import matplotlib.pyplot as plt
import numpy as np

# Titik-titik utama (koordinat 2D)
points = {
    '0': np.array([0, 0]),
    '1': np.array([2, 2]),
    '2': np.array([4, 0]),
    '3': np.array([2, -2]),
    '4': np.array([1, 3]),
    '5': np.array([5, -2]),
    'h': np.array([3, 0])
}

# Gambar segitiga besar (0-1-2-3-0)
fig, ax = plt.subplots(figsize=(8, 6))

# Gambar garis-garis poligon
polygon_pts = ['0', '1', '2', '3', '0']
for i in range(len(polygon_pts)-1):
    p1 = points[polygon_pts[i]]
    p2 = points[polygon_pts[i+1]]
    ax.plot([p1[0], p2[0]], [p1[1], p2[1]], 'k')

# Garis hijau 4–5
p4, p5 = points['4'], points['5']
ax.plot([p4[0], p5[0]], [p4[1], p5[1]], 'g', linewidth=2)

# Vektor V0h (merah)
p0, ph = points['0'], points['h']
ax.plot([p0[0], ph[0]], [p0[1], ph[1]], 'r-', linewidth=2)
ax.text((p0[0]+ph[0])/2, (p0[1]+ph[1])/2 + 0.2, 'V0h', color='r')

# Vektor vp dan up (dashed, merah)
vp = np.array([1.8, 1.5])
up = np.array([1.8, -1.5])
ax.plot([p0[0], vp[0]], [p0[1], vp[1]], 'r--')
ax.plot([p0[0], up[0]], [p0[1], up[1]], 'r--')
ax.text(vp[0], vp[1]+0.2, 'vp', color='r')
ax.text(up[0], up[1]-0.2, 'up', color='r')

# Vektor û, v̂, n̂ (biru)
u_hat = np.array([1, 0])
v_hat = np.array([0.5, 1])
n_hat = np.array([0, 1.5])
ax.arrow(p0[0], p0[1], u_hat[0], u_hat[1], color='b', head_width=0.1)
ax.arrow(p0[0], p0[1], v_hat[0], v_hat[1], color='b', head_width=0.1)
ax.arrow(p0[0], p0[1], n_hat[0], n_hat[1], color='b', head_width=0.1)
ax.text(u_hat[0]+0.1, u_hat[1], 'û', color='b')
ax.text(v_hat[0], v_hat[1]+0.1, 'v̂', color='b')
ax.text(n_hat[0]-0.2, n_hat[1]+0.1, 'n̂', color='b')

# Tandai semua titik
for label, coord in points.items():
    ax.plot(coord[0], coord[1], 'ko')
    ax.text(coord[0]+0.1, coord[1]+0.1, label)

# Atur tampilan
ax.set_aspect('equal')
ax.set_xlim(-1, 6)
ax.set_ylim(-3, 4)
ax.axis('off')
plt.title("Out-of-bounds geometry (Replika)")
plt.show()
```
