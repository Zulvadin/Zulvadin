```
"""
DATAPLOTS - Compression Test Results with Volume Regression Analysis
"""

import matplotlib.pyplot as plt
import numpy as np

# Set up plot area
plt.axis([0, 140, 0, 100])
plt.axis('on')
plt.grid(True)

# Data
t = [20, 40, 60, 80, 100, 120]  # time (hrs)
T = [30, 35, 43, 55, 70, 85]    # temperature (K)
p = [2, 3, 4, 5.3, 7.3, 9.6]     # pressure (psi)
v = [0.6, 0.58, 0.54, 0.46, 0.35, 0.2]  # volume (cns)

# Scale data for plotting
pp = [pressure * 10 for pressure in p]  # scale pressure by 10
vv = [volume * 100 for volume in v]     # scale volume by 100

# Plot data
plt.plot(t, T, color='r', label='Temperature', marker='o')
plt.plot(t, pp, color='b', label='Pressure', marker='s')
plt.plot(t, vv, color='g', label='Volume', marker='d')
plt.legend(loc='upper left')

# Add pressure scale on left
for y in np.arange(0, 101, 20):
    s = y / 10
    plt.text(-10, y, f"{s:.1f}", color='b', ha='right', va='center')

# Add volume scale on right
for y in np.arange(0, 101, 20):
    s = y / 100
    plt.text(163, y, f"{s:.2f}", color='g', ha='left', va='center')
    plt.text(153, y, '_', color='g')

# Add volume scale ticks
for y in np.arange(0, 100, 3):
    plt.text(157, y, '-', color='g')

# Labels
plt.xlabel('Time (hrs)')
plt.ylabel('Temperature (K)', color='r')
plt.text(151, 65, "Pressure (psi)", rotation=90, color='b')
plt.text(170, 65, "Volume (cns)", rotation=90, color='g')

# Title
plt.title('Compression Test Results')

# ---- Regression analysis for Volume v vs t ----
n = len(v)

# Calculate averages
c1 = np.sum(t)/n  # average of t
c2 = np.sum(v)/n  # average of v (using original v values)

# Calculate sums for regression
a = np.multiply(t, v)  # t*v element-wise
c3 = np.sum(a)         # sum(t*v)

a = np.multiply(t, t)  # t*t element-wise
c4 = np.sum(a)         # sum(t²)

# Calculate line parameters A and B
A = (c3 - n*c1*c2)/(c4 - n*c1*c1)
B = c2 - A*c1

# Plot regression line with scatter dots
for tp in np.arange(t[0], t[-1]+1, 2):  # tp from first to last t value in steps of 2
    vp = A*tp + B
    vp = vp * 100  # scale vp for plotting (same as vv scaling)
    plt.scatter(tp, vp, color='g', s=1, alpha=0.5)

# Calculate RMS error
sumse = 0
for i in range(len(t)):
    e = (v[i] - (A*t[i] + B))
    ee = e*e
    sumse = sumse + ee
rms = np.sqrt(sumse/n)

# Regression equation and values display
plt.text(60, 28, 'v = At + B', color='g')
plt.arrow(78, 30, 6, 6, head_length=3, head_width=1.5, color='g', linewidth=0.5)

# Display start and end values of regression line
vp1 = A*t[0] + B
vp1 = '%7.4f' % vp1
plt.text(2, 64, vp1, color='g')

vp2 = A*t[-1] + B
vp2 = '%7.4f' % vp2
plt.text(122, 25, vp2, color='g')

# Display regression coefficients
Ap = '%7.5f' % A
plt.text(65, 18, 'A =', color='g')
plt.text(72, 18, Ap, color='g')

Bp = '%7.5f' % B
plt.text(65, 12, 'B =', color='g')
plt.text(73, 12, Bp, color='g')

# Display RMS error
rms_str = '%7.3f' % rms
plt.text(95, 3, 'RMS error =', color='g')
plt.text(123, 3, rms_str, color='g')

plt.show()
```
